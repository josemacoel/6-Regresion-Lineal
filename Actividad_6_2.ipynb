### **Limpieza Inicial**

Se realizó una limpieza final del archivo. Algunas columnas encontraban en formato str y se obtuvo un promedio para aquellos valores iguales o mayor a 1, para posteriormente cambiar los valores a float y no modificar el tamaño de la columna o los datos. 

```python
(Gir["host_acceptance_rate"] == "No Data Available").sum()
```

```python
# Eliminar los símbolos de porcentaje
Gir['host_acceptance_rate'] = Gir['host_acceptance_rate'].str.replace('%', '')
Gir['host_acceptance_rate'] = Gir['host_acceptance_rate'].str.replace('No Data Available', "91.5")

# Convertir la columna a tipo float
Gir['host_acceptance_rate'] = Gir['host_acceptance_rate'].astype(float)
Gir['host_acceptance_rate'].fillna(91.5, inplace=True)
```

```python
promedio = Gir.loc[Gir['host_acceptance_rate'] >= 1, 'host_acceptance_rate'].mean()
promedio
```

```python
(Gir["host_response_rate"] == "No Data Available").sum()
```

```python
# Eliminar los símbolos de porcentaje
Gir['host_response_rate'] = Gir['host_response_rate'].str.replace('%', '')
Gir['host_response_rate'] = Gir['host_response_rate'].str.replace('No Data Available', "96")

# Convertir la columna a tipo float
Gir['host_response_rate'] = Gir['host_response_rate'].astype(float)
Gir['host_response_rate'].fillna(96, inplace=True)
```
